Q2: What is a scatter plot? Use the following code to generate data for x and y. Using this generated data 
plot a scatter plot.
Note: Also add title, xlabel, and ylabel to the plot.

```python
""" A scatter plot is a set of points plotted on a horizontal and vertical axes."""
```

```python
import numpy as np

np.random.seed(3)

x = 3 + np.random.normal(0, 2, 50)

y = 3 + np.random.normal(0, 2, len(x))
```

```python
import matplotlib.pyplot as plt
plt.scatter(x,y)
plt.xlabel("this is my x axis")
plt.ylabel("this is my y axis ")
plt.title("this is my scatter plot ")
```

Q3: Why is the subplot() function used? Draw four line plots using the subplot() function.

Use the following data:

```python
""" subplot divides the current figure into rectangular panes that are numbered row-wise. Each pane contains an axes. Subsequent plots are output to the current pane. subplot(m,n,p) creates an axes in the p -th pane of a figure divided into an m -by- n matrix of rectangular panes."""
```

```python
import numpy as np
import matplotlib.pyplot as plt

fig,axes = plt.subplots(2, 2) 

x = np.array([0, 1, 2, 3, 4, 5])
y = np.array([0, 100, 200, 300, 400, 500])
y1 = np.array([50, 20, 40, 20, 60, 70])
y2 = np.array([10, 20, 30, 40, 50, 60])
y3 = np.array([200, 350, 250, 550, 450, 150])

axes[0,0].plot(x,y)
axes[0,1].plot(x,y1)
axes[1,0].plot(x,y2)
axes[1,1].plot(x,y3)
```

Q4: What is a bar plot? Why is it used? Using the following data plot a bar plot and a horizontal bar plot. 

```python
#It allows you to compare different sets of data among different groups easily. It instantly demonstrates this relationship using two axes, where the categories are on one axis and the various values are on the other. 
#A bar graph can also illustrate important changes in data throughout a period of time.
```

```python
#Box plots are used to show distributions of numeric data values, especially when you want to compare them between multiple groups. 
#They are built to provide high-level information at a glance, offering general information about a group of data's symmetry, skew, variance, and outliers.
```

```python
import matplotlib.pyplot as plt

import numpy as np

box1 = np.random.normal(100, 10, 200)

box2 = np.random.normal(90, 20, 200)

plt.boxplot([box1,box2])
plt.show()
```
