```python
"""
1. scatterplot: The relationship between `x` and `y` can be shown for different subsets
2. displot: can get easily where the data is highly distributed
3. relplot: The relationship between `x` and `y` can be shown for different subsets of the data using the `hue`, `size`, and `style` parameters
hue : Grouping variable that will produce elements with different colors
size: Grouping variable that will produce elements with different sizes
style: Grouping variable that will produce elements with different styles
4. catplot:This function provides access to several axes-level functions that show the relationship between a numerical and one or more categorical
variables using one of several visual representations.
5. pairplot: By default, this function will create a grid of Axes such that each numeric variable in ``data`` will by shared across the y-axes across a single row and
the x-axes across a single column. The diagonal plots are treated differently: a univariate distribution plot is drawn to show the marginal
distribution of the data in each column.
"""
```

```python
import seaborn as sns
fmri=sns.load_dataset("fmri")
```

```python
fmri
```

```python
event_stim=fmri.query("event=='stim'")
event_cue=fmri.query("event=='cue'")
region_parietal=fmri.query("region=='parietal'")
region_frontal=fmri.query("region=='frontal'")
```

```python
sns.lineplot(x=event_stim.timepoint,y=event_stim.signal)
```
